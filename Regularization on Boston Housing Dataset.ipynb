```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
#from the above discription of Columns :
#inputs are [CRIM, ZN, INDUS, CHAS, NOX, RM, AGE, DIS, RAD, TAX, PTRATIO, B, LSTAT]
#output is [MEDV]
```

```python
# to load dataset
df=pd.read_csv('BostonHousing.csv')
df.head()
```

```python
#selecting input and output from dataset df_new
x=df.drop("medv",axis=1)  #select input
y=df["medv"]  #select outp[ut]
```

```python
x.head()
```

```python
#train data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#training model with 70% data
#using linearregresion cause price is cont numeric
#multiple linear regression cause multiple input
from sklearn.linear_model import LinearRegression
#creating an object 
lr=LinearRegression()
```

```python
#training model wityh 70% data
lr.fit(x_train,y_train)
```

```python
#check slope and intercept of each input
slope=lr.coef_
#print(slope)
slope=[np.round(i) for i in slope]
print(slope)
```

```python
x_train.columns
```

```python
#checking slope of all cols
dict={"input":x_train.columns,"slope":slope}
df1=pd.DataFrame(dict)
df1
```

```python
# check intercept
lr.intercept_
```

```python
#check training score inbuilt method score() which is defined in inbuilt class linearregression
```

```python
#score always in between 0-1 closer to 1 is good closet to 0 is bad
print("training score",lr.score(x_train,y_train))
```

```python
#test the model with 30% data
y_pred=lr.predict(x_test)
y_pred
```

```python
#or using r2score
from sklearn.metrics import r2_score
print("testing score",r2_score(y_test,y_pred))
```

# training score is 71 and testing score is 78 means our model is overfit ,training score>testing score now we have to add error on training time means apply regularisation

```python
#2 types of regularisation
#1 lasso(l1)
#2 ridge(l2)
''' Add some error during training time 
   1. Lasso Regularization : L1  
   RSS+lambda*sum(abs(all coefficient)) here lambda is also known as 
   hyper-parameter, value of lambda  can be 200,400,600,..1000
   2. Ridge Regularization : L2
   RSS+lambda*sum(square(all coefficient)) here lambda is also known as 
   hyper-parameter, value of lambda  can be 0 to 50
    '''
```

```python
#lasso regularisation means l1 call inbuilt class Lasso into sklearn lib
from sklearn.linear_model import Lasso
l1=Lasso(400)
```

```python
#train the model with lasso,use inbuilt method fit() of lasso class
l1.fit(x_train,y_train)
```

```python
#check the training score after train the model with lasso
l1.score(x_train,y_train)
```

```python
#select the right value of lambda/alpha for lasso
for i in range(200,1001,50):     #start=200 stop=1001 staep=5
    #first time lambda=200,next time=250...
    #create object of lasso
    l1=Lasso(i)  #here i=200 first time
    #train the model with l1 object
    l1.fit(x_train,y_train)
    #check training score
    train_score=l1.score(x_train,y_train)
    #to find testing score
    test_score=l1.score(x_test,y_test)
    print("value of lambda: {}\n training score:{}\n testing score:{}".
         format(i,train_score,test_score))
    print("\n")
```

```python
#call ridge class
from sklearn.linear_model import Ridge
#select right value of lambda for ridge regularisation
#create an object
for i in range(0,51,1):  #here i is value of lambda
    l2=Ridge(i)   #first time i=0 then i=1...
     #train the model with l1 object
    l2.fit(x_train,y_train)
    #check training score
    s1=l2.score(x_train,y_train)
    #to find testing score
    s2=l2.score(x_test,y_test)
    print("value of lambda: {}\n training score:{}\n testing score:{}".
         format(i,s1,s2))
    print("\n")
    
```
